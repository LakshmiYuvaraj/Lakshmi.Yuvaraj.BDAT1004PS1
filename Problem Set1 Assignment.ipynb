Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area 
of a triangle with side lengths a, b, and c is
s(s - a)(s -b)(s -c)
, where 
s = (a+b+c)/2.

https://www.geeksforgeeks.org/c-program-find-area-triangle/     #used below code from this link

```python
using System;                                          

namespace Question_3
{
    class Program
    {
        static void Main(string[] args)
        {
            
            static float findArea(float a, float b, float c)
            {

                // Length of sides must be positive and sum of any two sides
                // must be smaller than third side.
                if (a < 0 || b < 0 || c < 0 ||
                (a + b <= c) || a + c <= b ||
                    b + c <= a)
                {
                    Console.Write("Not a valid triangle");
                    System.Environment.Exit(0);
                }
                float s = (a + b + c) / 2;
                return (float)Math.Sqrt(s * (s - a) *
                                    (s - b) * (s - c));
            }

            
            {
                float a = 2.0f;
                float b = 2.0f;
                float c = 2.0f;

                Console.Write("Area is " + findArea(a, b, c));
            }
        }
    }
}
```

# Question 5

a. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False 
depending on whether the point (x,y) lies in the rectangle with lower left 
corner (x1,y1) and upper right corner (x2,y2).

```python
using System;

namespace Question_5
{
    class Program
    {
        static void Main(string[] args)
        {
            static bool FindPoint(int x1, int y1, int x2,
                      int y2, int x, int y)
            {
                if (x > x1 && x < x2 &&
                    y > y1 && y < y2)
                    return true;

                return false;
            }

            {

                // bottom-left and top-right
                // corners of rectangle
                int x1 = 0, y1 = 0,
                    x2 = 2, y2 = 3;

                // given point
                int x = 1, y = 1;

                // function call
                if (FindPoint(x1, y1, x2, y2, x, y))
                    Console.Write("True");
                else
                    Console.Write("False");
            }
        }
    }
}
```
